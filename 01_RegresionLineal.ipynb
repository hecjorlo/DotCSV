```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos la libreria
boston = load_boston()
print(boston.DESCR)
```

```python
X = np.array(boston.data[:,5]) # Columna Rooms
Y = np.array(boston.target) # N medio de hab

plt.scatter(X,Y, alpha=0.3)
plt.show()
```

```python
# Añadimos columna de 1 para termino independiente y hacemos las transpuesta
X_mod = np.array([np.ones(506), X]).T
print(X_mod)
```

Formula minimizar minimizar error cuadratico (MSE): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
B = np.linalg.inv(X_mod.T @ X_mod) @ X_mod.T @ Y
B
```

Nuestra regresion lineal sera: y = ax +b

```python
a=B[1]
b=B[0]
```

```python
plt.plot([4,9],[a*4+b,a*9+b], c='red')
plt.scatter(X,Y, alpha=0.3)
plt.show()
```
