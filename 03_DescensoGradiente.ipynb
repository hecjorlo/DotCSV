```python
import numpy as np
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
```

```python
func = lambda z: np.sin(1/2 * z[0] **2 -1/4 * z[1] **2 +3) * np.cos(2*z[0] +1 - np.e**z[1])
```

```python
func([5,3])
```

```python
X = np.linspace(-2,2,100)
Y = np.linspace(-2,2,100)
Z = np.zeros((100,100))

for ix,x in enumerate(X):
  for iy,y in enumerate(X):
    Z[iy,ix] = func([x,y])

print(Z)
```

```python
plt.contourf(X,Y,Z,100)
plt.colorbar()
plt.show()
```

```python
#Numero aleatorio entre [-2,2]
np.random.seed(42)
Theta = np.random.rand(2) * 4 - 2

plt.contourf(X,Y,Z,100)
plt.colorbar()
plt.plot(Theta[0],Theta[1],'o',c='red')
plt.show()
```
